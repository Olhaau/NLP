```python
# open browser
options = webdriver.ChromeOptions()
options.add_argument('--lang=en-EN')
#options.add_argument('--proxy-server=207.91.10.234')
#192.240.46.123
#options.add_experimental_option('prefs', {'intl.accept_languages': 'en,en_US'})
driver = webdriver.Chrome(options=options)
```

```python
topic = "EU green deal"

driver.get(
    'https://www.google.com/search?q=' 
    + topic.replace(' ', '+') 
    + '&source=lnms&tbm=nws'
)
```

```python
#close dsgvo message
driver.find_element_by_xpath('//*[contains(text(), "' + "I agree" + '")]').click()
```

```python
n_pages = 10

articles = []

for k in range(n_pages - 1):
    for i in range(len(driver.find_elements_by_xpath('//div[@id="rso"]/div/g-card[1]/div/div/a/div/div[2]/div[2]'))):
        #print("- Artikel: {} ----------------".format(i + 1))
        date_time = driver.find_elements_by_xpath(
            '//div[@id="rso"]/div/g-card[1]/div/div/a/div/div[2]/div[4]/p/span')[i].text
        source = driver.find_elements_by_xpath(
            '//div[@id="rso"]/div/g-card[1]/div/div/a/div/div[2]/div[1]/span')[i].text
        title = driver.find_elements_by_xpath('//div[@id="rso"]/div/g-card[1]/div/div/a/div/div[2]/div[2]')[i].text
        desc = driver.find_elements_by_xpath('//div[@id="rso"]/div/g-card[1]/div/div/a/div/div[2]/div[3]')[i].text
        url_art = driver.find_elements_by_xpath('//div[@id="rso"]/div/g-card[1]/div/div/a')[i].get_attribute('href')
        articles.append(
            {
            "title": title,
            "source": source,
            "description": desc,
            "url": url_art,
            "date_time": date_time
            }
        )
    driver.find_element_by_id("pnnext").click()
    sleep(5)
    #print(date_time)
    #print(source)
    #print(title)
    #print(desc)
    #print(url_art)

articles_data = pd.DataFrame(articles)
articles_data
```

```python
articles_data
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
#try to get the contents of the pages
htmls, contents = [], []
for art_url in articles_data.url:
    try:
        req = requests.get(art_url)
        soup = BeautifulSoup(req.content, 'html.parser')
        htmls.append(soup)
    except: htmls.append("")
    try:
        els = soup.find_all("p")
    except: els = []
    txt = ""
    for el in els:
        if len(el.text) >= 100:
            txt += "\n\n" + el.text.replace("\n", "").replace("  ","")
    contents.append(txt)
```

```python
articles_data["html"] = htmls
articles_data["content"] = contents
```

```python
# detect its language
from langdetect import detect

def detector(x):
    try:
       return detect(x)
    except:
        None 
```

```python
vader_scores = [analyzer.polarity_scores(art)["compound"] for art in articles_data.content]
#rem: 
# pos sentiment: compound score >= 0.05
# neutral sentiment: -0.05 < compound score < 0.05
# neg sentiment: compound score <= -0.05
```

```python
art_expl = articles_data.content[4]
detector("")
```

```python
articles_data["vader_sc"] = vader_scores
```

```python
langs = [detector(art) for art in articles_data.content]
articles_data["language"] = langs
```
